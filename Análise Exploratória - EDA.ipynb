```python
# Análise Exploratória dos Dados (EDA)
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projeto Semantix/global_energy_consumption.csv')

print('Verificação Inicial: \n')
print(df.info())
```

```python
# Verificar os primeiros registros
print(df.head().to_string())

# Verificar os últimos registros
print(df.tail().to_string())
```

```python
# Verificar quantidade de linhas e colunas
print('Quantidade: ', df.shape)

# Checar o tipo dos dados
print('Tipos de dados:\n', df.dtypes)

# Checar valores nulos
print('Valores nulos:\n', df.isnull().sum())

# Checar dados duplicados
print('Análise de dados duplicados:\n', df.duplicated().sum())

# Checar dados únicos
print('Análise de dados únicos:\n', df.nunique())

# Estatísticas
print('Estatísticas dos dados:\n', df.describe())
```

```python
# Mostrar o nome dos países únicos
paises_unicos = df['Country'].unique()
for pais in paises_unicos:
    print(pais)
```

```python
# Calcular o consumo total de energia para cada país ao longo de todos os anos
# Agrupamos por 'Country' e somamos a coluna 'Total Energy Consumption (TWh)'
consumo_total_por_pais = df.groupby('Country')['Total Energy Consumption (TWh)'].sum()

# Converter o resultado para um DataFrame
consumo_total_por_pais_df = consumo_total_por_pais.reset_index()

# Ordenar o DataFrame pelo 'Total Energy Consumption (TWh)' em ordem decrescente
ranking_consumo = consumo_total_por_pais_df.sort_values(
    by='Total Energy Consumption (TWh)',
    ascending=False)

# Exibir o ranking
print("Ranking dos Países por Total Energy Consumption (TWh): ")
print(ranking_consumo)
```

```python
# Calcular a média de dependência de combustível fóssil para cada país ao longo de todos os anos
# Agrupamos por 'Country' e pegamos a média da coluna 'Fossil Fuel Dependency (%)'
media_dependencia_por_pais = df.groupby('Country')['Fossil Fuel Dependency (%)'].mean()

# Converter o resultado para um DataFrame
media_dependencia_por_pais_df = media_dependencia_por_pais.reset_index()

# Ordenar o DataFrame pelo 'Fossil Fuel Dependency (%)' em ordem decrescente
ranking_dependencia = media_dependencia_por_pais_df.sort_values(
    by='Fossil Fuel Dependency (%)',
    ascending=False)

# Exibir o ranking
print("Ranking dos Países por Média de Dependência de Combustível Fósil: ")
print(ranking_dependencia)
```

```python
# Calcular a emissão de carbono para cada país ao longo de todos os anos
# Agrupamos por 'Country' e somamos a coluna 'Carbon Emissions (Million Tons)'
emissao_por_pais = df.groupby('Country')['Carbon Emissions (Million Tons)'].sum()

# Converter o resultado para um DataFrame
emissao_por_pais_df = emissao_por_pais.reset_index()

# Ordenar o DataFrame pela 'Carbon Emissions (Million Tons)' em ordem decrescente
ranking_emissao = emissao_por_pais_df.sort_values(
    by='Carbon Emissions (Million Tons)',
    ascending=False)

# Exibir o ranking
print("Ranking dos Países por Emissão de Carbono: ")
print(ranking_emissao)
```

```python
# Analisar um país individualmente
# Definindo o país que será analisado
pais_selecionado = 'USA'
dados_pais = df[df['Country'] == pais_selecionado]

# Calcular a média de consumo de energia por ano para o país selecionado
media_consumo_pais_por_ano = dados_pais.groupby('Year')['Total Energy Consumption (TWh)'].mean().reset_index()
media_consumo_pais_por_ano_round = round(media_consumo_pais_por_ano, 2)

# Imprimir o resultado
print(f"Média de Consumo de Energia para '{pais_selecionado}' por Ano:")
print(media_consumo_pais_por_ano_round)
```

```python
# Encontrar o ano com a maior média de consumo
ano_maior_consumo_idx = media_consumo_pais_por_ano['Total Energy Consumption (TWh)'].idxmax()
ano_maior_consumo_info = media_consumo_pais_por_ano.loc[ano_maior_consumo_idx]

ano_maior = ano_maior_consumo_info['Year']
maior_media = ano_maior_consumo_info['Total Energy Consumption (TWh)']

print(f"\nO ano com a MAIOR média de consumo para {pais_selecionado} foi: {int(ano_maior)} com {maior_media:.2f}")

# Encontrar o ano com a menor média de consumo
ano_menor_consumo_idx = media_consumo_pais_por_ano['Total Energy Consumption (TWh)'].idxmin()
ano_menor_consumo_info = media_consumo_pais_por_ano.loc[ano_menor_consumo_idx]

ano_menor = ano_menor_consumo_info['Year']
menor_media = ano_menor_consumo_info['Total Energy Consumption (TWh)']

print(f"\nO ano com a MENOR média de consumo para {pais_selecionado} foi: {int(ano_menor)} com {menor_media:.2f}")
```

```python
# Calcular a média de emissão de gás carbônico por ano para o país selecionado
media_emissao_carbono_por_ano = dados_pais.groupby('Year')['Carbon Emissions (Million Tons)'].mean().reset_index()
media_emissao_carbono_por_ano_round = round(media_emissao_carbono_por_ano, 2)

# Imprimir o resultado
print(f"Média de Emissão de Carbono para '{pais_selecionado}' por Ano:")
print(media_emissao_carbono_por_ano_round)
```

```python
# Encontrar o ano com a maior média de emissão de gás carbônico
ano_maior_emissao_idx = media_emissao_carbono_por_ano['Carbon Emissions (Million Tons)'].idxmax()
ano_maior_emissao_info = media_emissao_carbono_por_ano.loc[ano_maior_emissao_idx]

ano_maior_ec = ano_maior_emissao_info['Year']
maior_media_ec = ano_maior_emissao_info['Carbon Emissions (Million Tons)']

print(f"\nO ano com a MAIOR média de emissão de gás carbônico para {pais_selecionado} foi: {int(ano_maior_ec)} com {maior_media_ec:.2f}")

# Encontrar o ano com a menor média de emissão de gás carbônico
ano_menor_emissao_idx = media_emissao_carbono_por_ano['Carbon Emissions (Million Tons)'].idxmin()
ano_menor_emissao_info = media_emissao_carbono_por_ano.loc[ano_menor_emissao_idx]

ano_menor_ec = ano_menor_emissao_info['Year']
menor_media_ec = ano_menor_emissao_info['Carbon Emissions (Million Tons)']

print(f"\nO ano com a MENOR média de emissão de gás carbônico para {pais_selecionado} foi: {int(ano_menor_ec)} com {menor_media_ec:.2f}")
```

```python
# Calcular a média de dependência de combustível fóssil por ano para o país selecionado
media_dependencia_fossil = dados_pais.groupby('Year')['Fossil Fuel Dependency (%)'].mean().reset_index()
media_dependencia_fossil_round = round(media_dependencia_fossil, 2)

# Imprimir o resultado
print(f"Média de Dependência de Combustível Fósil para '{pais_selecionado}' por Ano:")
print(media_dependencia_fossil)
```

```python
# Encontrar o ano com a maior média de dependência de combustível fóssil
ano_maior_dependencia_fossil_idx = media_dependencia_fossil['Fossil Fuel Dependency (%)'].idxmax()
ano_maior_dependencia_fossil_info = media_dependencia_fossil.loc[ano_maior_dependencia_fossil_idx]

ano_maior_fd = ano_maior_dependencia_fossil_info['Year']
maior_media_fd = ano_maior_dependencia_fossil_info['Fossil Fuel Dependency (%)']

print(f"\nO ano com a MAIOR média de dependência de combustível fóssil para {pais_selecionado} foi: {int(ano_maior_fd)} com {maior_media_fd:.2f}")

# Encontrar o ano com a menor média de dependência de combustível fóssil
ano_menor_dependencia_fossil_idx = media_dependencia_fossil['Fossil Fuel Dependency (%)'].idxmin()
ano_menor_dependencia_fossil_info = media_dependencia_fossil.loc[ano_menor_dependencia_fossil_idx]

ano_menor_fd = ano_menor_dependencia_fossil_info['Year']
menor_media_fd = ano_menor_dependencia_fossil_info['Fossil Fuel Dependency (%)']

print(f"\nO ano com a MENOR média de dependência de combustível fóssil para {pais_selecionado} foi: {int(ano_menor_fd)} com {menor_media_fd:.2f}")
```
